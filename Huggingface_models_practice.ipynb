```python
import torch
from transformers import pipeline
```

```python
translator=pipeline(task="translation",
                    model="facebook/nllb-200-distilled-600M",
                    torch_dtype=torch.bfloat16
                    )
```

```python
text="""\
My puppy is adorable, \
Your kitten s cute.
Her panda is friendly.
His llama is thoughful.\
We all have nice pets!"""
```

```python
text_translated=translator(text,
                          src_lang="eng_Latn",
                          tgt_lang="fra_Latn")
```

```python
text_translated
```

```python
import gc
del translator
gc.collect()
```

```python
summarizer=pipeline(task="summarization",
                    model="facebook/bart-large-cnn",
                    torch_dtype=torch.bfloat16)
```

```python
text="""
Accomplished Lead Statistical Programmer with a passion for delivering valuable data through statistical
and data retrieval methodologies. Over 10 years of experience including clinical trial data analysis in
 Oncology, Women’s health, Infectious disease, Blood Pressure and Diabetes related therapeutic areas.
 Highly skilled in data analysis, data visualization with hands on knowledge of machine learning/deep
 learning. Seeking to learn new ML skills and applying knowledge to dynamic, growth-oriented role.
 Working as Lead-Statistical Programmer in multiple phases of clinical trials for women’s health related
  therapeutic area. Leading a team of 9 junior and senior programmers.
"""
```

```python
summary=summarizer(text,
                   min_length=10,
                   max_length=100
                   )
```
